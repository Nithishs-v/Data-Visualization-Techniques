```python
import seaborn
import matplotlib.pyplot as plt
seaborn.set(style = 'whitegrid')
tip = seaborn.load_dataset("tips")
print(tip)
seaborn.stripplot(x="day", y="total_bill", data=tip)
plt.show()
```

```python
seaborn.stripplot(x="day", y="total_bill", data=tip, jitter=0.99)
```

```python
seaborn.stripplot(y="total_bill", x="day", data=tip, linewidth=1,edgecolor='red')
```

```python
seaborn.stripplot(x="sex", y="total_bill", hue="day", data=tip)
```

```python
seaborn.stripplot(x="day", y="total_bill", hue="smoker",
data=tip, palette="Set2", dodge=True)
```

```python
import seaborn as sns
sns.boxplot(x = 'day', y = 'total_bill', data = tip)
```

```python
seaborn.swarmplot(x='day', y='tip', data=tip)
```

```python
seaborn.swarmplot(x="day", y="total_bill", hue="time", data=tip,size=10)
```

```python
data = sns.load_dataset("mpg")
sns.jointplot(x = "mpg", y = "acceleration",
kind = "scatter", data = data)
plt.show()
```

```python
# loading dataset
data = sns.load_dataset("attention")
# draw jointplot with
# hex kind
sns.jointplot(x = "solutions", y = "score",
kind = "hex", data = data)

# show the plot
plt.show()
```

```python
data = sns.load_dataset("exercise")
sns.jointplot(x = "id", y = "pulse",
kind = "kde", data = data)
plt.show()
```

```python
data = sns.load_dataset("titanic")
sns.jointplot(x = "age", y = "fare",
kind = "reg", data = data,
dropna = True)
plt.show()
```
