**DIRECT PLOTTINGS**

Pandas is a Python library with data frame features that supplies built-
in options for plotting visualizations in a two-dimensional tabular style.

```python
import pandas as pd
dataset = pd.read_csv("/content/drive/MyDrive/DATASET/salary.csv")
rank = dataset['rank']
discipline = dataset['discipline']
phd = dataset['phd']
sevice = dataset['sevice']
sex = dataset['sex']
salary = dataset['salary']
dataset.head()
```

**Line Plot**

```python
dataset[["rank", "discipline","phd","sevice", "sex","salary"]].plot()
```

```python
dataset[["phd","sevice"]].plot()
```

```python
dataset1 = dataset.groupby(['sevice']).sum()
dataset1.sort_values("salary", ascending = False,
inplace=True)
dataset1.head()
```

```python
dataset1["salary"].plot.bar()
```

**Bar Plot**

```python
dataset[[ 'phd', 'sevice' ]].head(10).plot.bar()
```

```python
dataset[['phd', 'sevice']].head(10).plot.bar(title="Ph.D. Vs Service\n 2018")
```

```python
dataset[['phd', 'sevice']].head(10).plot.bar (title="Ph.D. Vs Service\n 2018" , color=['g','red'])
```

**Pie Chart**

```python
dataset["salary"].head(10).plot.pie(autopct='%.2f')
```

**Box Plot**

```python
dataset[["phd","salary"]].head(100).plot.box()
```

```python
dataset[["phd","sevice"]].plot.box()
```

**Histogram Plot**

A histogram can be used to represent a specific variable or set ofvariables.

```python
dataset["salary"].head(20).plot.hist()
```

**Scatter Plot**

A scatter plot shows the relationship between two factors of an experiment
(e.g. phd and service).

```python
import pandas as pd
import numpy as np
dataset = pd.read_csv("/content/drive/MyDrive/DATASET/salary.csv")
dataset.plot(kind='scatter', x='phd', y='sevice',title='Popuation vs area and density\n 2018', s=0.9)
```

**SEABORN**

The Python Seaborn library provides various plotting representations

visualizing data.

 A strip plot is a scatter plot where one of the variables
is categorical.

Strip plots can be combined with other plots to provide
additional information.

**Strip Plot**

uses strip plotting to display data per salary category.

```python
import seaborn as sns
sns.stripplot( x =dataset['salary'])
```

```python
sns.stripplot( x = dataset['sex'], y= dataset['salary'],data=dataset);
```

```python
sns.stripplot( x = dataset['discipline'], y =dataset['salary'], data=dataset, jitter=1)
```

```python
sns.stripplot( x= dataset['salary'], y = dataset['discipline'],data=dataset, jitter=True);
```

```python
sns.stripplot( x = dataset['rank'], y= dataset['salary'],data=dataset, jitter=True);
```

```python
sns.stripplot( x ='sex', y= 'salary', hue='rank',data=dataset, jitter=True )
```

**Box Plot**

```python
sns.boxplot(x = 'salary', y ='sex', data=dataset,whis=np.inf )
sns.stripplot(x = 'salary', y ='sex', data=dataset,jitter=True, color='0.02')
```

```python
sns.boxplot(x = 'rank', y = 'salary', data=dataset)
```

```python
sns.boxplot(x = dataset['salary'])
```

```python
sns.boxplot(x = 'rank', y = 'salary', hue='sex', data=dataset,palette='Set3')
```

```python
sns.boxplot(x = dataset['salary'], notch=True)
```

```python
sns.boxplot(x = dataset['salary'], whis=2)
```

```python
sns.boxplot(x = 'rank', y = 'salary', data=dataset)
sns.swarmplot(x = 'rank', y = 'salary', data=dataset,color='0.25')
```

**Swarm Plot**

A swarm plot is used to visualize different categories; it gives a clear
picture of a variable distribution against other variables.

```python
sns.swarmplot( x ='sex', y= 'salary', hue='rank', data=dataset,palette="Set2", dodge=True)
```

```python
sns.swarmplot( x ='sex', y= 'salary', hue='rank', data=dataset,palette="Set2", dodge=False)
```

**Joint Plot**

A joint plot combines more than one plot to visualize the selected patterns

```python
sns.jointplot(x = 'salary', y = 'sevice',data=dataset)
```

```python
import seaborn as sns
dataset = pd.read_csv("/content/drive/MyDrive/DATASET/salary.csv")
sns.jointplot('salary', 'sevice', data=dataset,kind='reg')
```

```python
sns.jointplot(data=dataset, x="salary", y="sevice",hue="sex",kind="kde")
```

```python
from scipy.stats import spearmanr
sns.jointplot('salary', 'sevice', data=dataset, stat_func=spearmanr )
```
