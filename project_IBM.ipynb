```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'GDP': [14000, 14300, 14500, 14700, 14900, 15100, 15200, 15400, 15600, 15800],
    'Recession_Flag': [True, True, False, False, True, True, False, False, False, False]  # Indicate recession years
}

# Create a DataFrame
df = pd.DataFrame(data)

# Separate data for recession and non-recession years
recession_df = df[df['Recession_Flag']]
non_recession_df = df[~df['Recession_Flag']]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot GDP during recession years
axes[0].plot(recession_df['Year'], recession_df['GDP'], marker='o', linestyle='-', color='red')
axes[0].set_title('GDP During Recession Years')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('GDP')
axes[0].grid(True)

# Plot GDP during non-recession years
axes[1].plot(non_recession_df['Year'], non_recession_df['GDP'], marker='o', linestyle='-', color='blue')
axes[1].set_title('GDP During Non-Recession Years')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('GDP')
axes[1].grid(True)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [1200, 1300, 1400, 1600, 1900, 2300, 2500, 2400, 2100, 1800, 1400, 1300],
    'Seasonality': [0.8, 0.7, 0.6, 0.5, 0.4, 0.6, 0.8, 0.9, 1.0, 0.9, 0.7, 0.6]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Normalize the seasonality values to determine bubble sizes
min_size = 100
max_size = 1000
df['Bubble_Size'] = df['Seasonality'].apply(lambda x: min_size + (max_size - min_size) * x)

# Create a bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Month'], s=df['Bubble_Size'], c=df['Sales'], cmap='coolwarm', alpha=0.7, edgecolors='k')

plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.colorbar(label='Sales Volume')
plt.grid(True)

# Customize the x-axis labels
plt.xticks(df.index, df['Month'])

# Show the plot
plt.show()
```

```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.express as px
from dash import Input, Output
import pandas as pd

# Sample data (replace with your actual data)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'GDP': [14000, 14300, 14500, 14700, 14900],
    'Unemployment_Rate': [8.2, 9.1, 9.5, 8.7, 7.8]
}

# Create a Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Recession Report Statistics", style={'textAlign': 'center'}),

    # Dropdowns (input container)
    dcc.Dropdown(
        id='statistic-dropdown',
        options=[
            {'label': 'GDP', 'value': 'GDP'},
            {'label': 'Unemployment Rate', 'value': 'Unemployment_Rate'}
        ],
        value='GDP',
        placeholder="Select a statistic",
        style={'width': '50%'}
    ),

    # Output Division (output container)
    dcc.Graph(id='output-plot'),
])

# Callback function to update the output container based on the selected statistic
@app.callback(
    Output('output-plot', 'figure'),
    Input('statistic-dropdown', 'value')
)
def update_output(selected_statistic):
    if selected_statistic == 'GDP':
        # Line plot of GDP
        return px.line(data, x='Year', y='GDP', title='GDP Over the Years')
    elif selected_statistic == 'Unemployment_Rate':
        # Bar plot of Unemployment Rate
        return px.bar(data, x='Year', y='Unemployment_Rate', title='Unemployment Rate Over the Years')
    else:
        return None

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.express as px
from dash import Input, Output
import pandas as pd

# Sample data (replace with your actual data)
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'GDP': [14000, 14300, 14500, 14700, 14900],
    'Unemployment_Rate': [8.2, 9.1, 9.5, 8.7, 7.8],
    'Inflation_Rate': [2.5, 3.0, 2.8, 2.2, 2.5]
}

# Create a Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Yearly Report Statistics", style={'textAlign': 'center'}),

    # Dropdowns (input container)
    dcc.Dropdown(
        id='statistic-dropdown',
        options=[
            {'label': 'GDP', 'value': 'GDP'},
            {'label': 'Unemployment Rate', 'value': 'Unemployment_Rate'},
            {'label': 'Inflation Rate', 'value': 'Inflation_Rate'}
        ],
        value='GDP',
        placeholder="Select a statistic",
        style={'width': '50%'}
    ),

    # Output Division (output container)
    dcc.Graph(id='output-plot'),
])

# Callback function to update the output container based on the selected statistic
@app.callback(
    Output('output-plot', 'figure'),
    Input('statistic-dropdown', 'value')
)
def update_output(selected_statistic):
    if selected_statistic == 'GDP':
        # Line plot of GDP
        return px.line(data, x='Year', y='GDP', title='GDP Over the Years')
    elif selected_statistic == 'Unemployment_Rate':
        # Bar plot of Unemployment Rate
        return px.bar(data, x='Year', y='Unemployment_Rate', title='Unemployment Rate Over the Years')
    elif selected_statistic == 'Inflation_Rate':
        # Line plot of Inflation Rate
        return px.line(data, x='Year', y='Inflation_Rate', title='Inflation Rate Over the Years')
    else:
        return None

if __name__ == '__main__':
    app.run_server(debug=True)
```
